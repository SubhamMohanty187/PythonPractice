##MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

Generally easy to get started for simple plots
Support for custom labels and texts
Great control of every element in a figure
High-quality output in many formats
Very customizable in general

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Simple Examples

x=np.arange(0,10)
y=np.arange(11,21)
a=np.arange(40,50)
b=np.arange(50,60)
```

### Scatter Plot

```python
plt.scatter(x,y, c = 'r')  # c = Color
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Graph')   # Title for the plot
plt.savefig('Test.jpg') # Saves the plot as image file
```

PLT Plot

```python
y = x*x
plt.plot(x,y, 'g*', linestyle='--', linewidth=2, markersize=10) # markersize is for dot size or star size
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

## Subplots|

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'b--')
```

## BarPlot

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x, y)
plt.bar(x2, y2, color = 'g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("histogram")
plt.show()
```

## BoxPlot in Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

## PieChart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'red', 'lightskyblue']
explode = (0, 0, 0.3, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)  # autopct means the format in which data will be reprsented in the piechart. here, it is float.

plt.title('PieChart')
plt.axis('equal')
plt.show()
```

#Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.dtypes
```

###Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
df = df.drop(columns=["species"]) # Dropping species column because species is a categorical variable and it can be used for correlaion.
df.corr()
```

```python
sns.heatmap(df.corr())
```

###JointPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=df, kind='hex')
```

```python
sns.jointplot(x='petal_length', y='petal_width', data=df, kind='reg')
```

###Pair plot
A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(data=df, hue='species')
```

###Dist plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['sepal_length'])
```

```python
sns.distplot(df['petal_length'],kde=False)
```

###Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

### BarPlot

```python
sns.barplot(data=df, x='sepal_length', y='species')
```

```python
sns.barplot(data=df, y='petal_width', x='species')
```

### Box plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(data=df, x='species', y='petal_length')
```

```python
sns.boxplot(data=df, hue='species', x='petal_width', palette='rainbow')
```

### Violin Plot
Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(data=df, hue='species', x='petal_width', palette='rainbow')
```
