```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country
```

Merging the two dataframes

```python
final_df = pd.merge(df,df_country, on='Country Code', how='left')
final_df
```

```python
final_df['Country'].value_counts()
```

```python
country_name = final_df['Country'].value_counts().index
country_val = final_df['Country'].value_counts().values
```

```python
## Piechart to plot Top # Countries using Zomato
plt.pie(x=country_val[:3], labels=country_name[:3], autopct="%1.2f%%")
```

#### Observation-: Zomato has maximum transaction in India followed by United States and UK.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()  ### size() is used to show the group by object.
```

```python
# To convert this to a dataframe, use reset_index()
ratings_df = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings_df
```

### Observations
1. 2148 people have given 0 rating or no ratings.
2. When Ratings between 4.5 to 4.9 --> Excellent
3. When Ratings between 4.0 to 4.4 --> Very Good
4. When Ratings between 3.5 to 3.9 --> Good
5. When Ratings between 2.5 to 3.4 --> Average
6. When Ratings between 1.8 to 2.4 --> Poor

```python
ratings_df.head()
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings_df, hue='Rating color', palette=['pink', 'red', 'orange', 'yellow', 'green', 'green'])
```

```python
sns.countplot(x='Rating color', data=ratings_df, hue='Rating color', palette=['pink', 'red', 'orange', 'yellow', 'green', 'green'])
```
